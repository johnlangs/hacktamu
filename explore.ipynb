```python
import datetime
import pprint
import requests 
```

```python
url = "http://localhost:4000/"
pp = pprint.PrettyPrinter(indent=4)
```

```python
flights = requests.api.get(url + "flights?date=2022-10-01").json()
pp.pprint(flights[0])
```

```python

for flight in flights:
    d = datetime.datetime.fromisoformat(flight['departureTime'])
    a = datetime.datetime.fromisoformat(flight['arrivalTime'])
    print(f"T1: {flight['duration']['locale']}, T2: {a - d}")
```
